```python
from sklearn.preprocessing import LabelEncoder

temp_df = df.copy()

temp_df = temp_df.drop("Loan_ID", axis=1)

encoder = LabelEncoder()

for col in temp_df.select_dtypes(include="object").columns:
    temp_df[col] = encoder.fit_transform(temp_df[col].astype(str))

plt.figure(figsize=(10,6))
sns.heatmap(temp_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
categorical_cols = [
    "Gender",
    "Married",
    "Dependents",
    "Self_Employed",
    "Credit_History"
]

for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
numerical_cols = [
    "LoanAmount",
    "Loan_Amount_Term"
]

for col in numerical_cols:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
import pandas as pd

df = pd.read_csv("../dataset/loan_prediction_cleaned.csv")
```

```python
# Features
X = df.drop("Loan_Status", axis=1)

# Target
y = df["Loan_Status"]

print("Features Shape:", X.shape)
print("Target Shape:", y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("Training Data:", X_train.shape)
print("Testing Data:", X_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_model = DecisionTreeClassifier(random_state=42)
```

```python
dt_model.fit(X_train, y_train)
```

```python
y_pred = dt_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)

print("Decision Tree Accuracy:", accuracy)
```

```python
train_accuracy = dt_model.score(X_train, y_train)

print("Training Accuracy:", train_accuracy)
```

```python
test_accuracy = dt_model.score(X_test, y_test)

print("Testing Accuracy:", test_accuracy)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

print(cm)
```

```python
plt.figure(figsize=(6,5))

sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues"
)

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Decision Tree Confusion Matrix")

plt.show()
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
import joblib

joblib.dump(dt_model, "../models/decision_tree.pkl")

print("Decision Tree model saved successfully!")
```

```python
loaded_model = joblib.load("../models/decision_tree.pkl")

sample_prediction = loaded_model.predict(X_test.iloc[:5])

print("Predictions:", sample_prediction)
```

```python
print("="*40)
print("Decision Tree Results")
print("="*40)

print(f"Training Accuracy : {train_accuracy:.4f}")
print(f"Testing Accuracy  : {test_accuracy:.4f}")
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)
```

```python
rf_train_accuracy = rf_model.score(X_train, y_train)

print("Training Accuracy:", rf_train_accuracy)
```

```python
rf_test_accuracy = rf_model.score(X_test, y_test)

print("Testing Accuracy:", rf_test_accuracy)
```

```python
plt.figure(figsize=(6,5))

sns.heatmap(
    rf_cm,
    annot=True,
    fmt="d",
    cmap="Greens"
)

plt.title("Random Forest Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.show()
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, rf_predictions))
```

```python
import joblib

joblib.dump(rf_model, "../models/random_forest.pkl")

print("Random Forest model saved successfully!")
```
